```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import sys
from scipy.interpolate import interp1d
```

```python
chi = 2
results_path = "./UFPV_chi2/"
cont = os.listdir(results_path)
matching_files = [f for f in cont if "e-01" in f and f.startswith("CH4_p01t") and f.endswith(".tout.kg")]
matching_files_sorted = sorted(matching_files)
print(matching_files_sorted)
z_st = 0.177
progvar_values = np.array([])
time_values =np.array([])
a =np.array([])
b=np.array([])
df = pd.DataFrame([], columns=['time', 'chi','c_st', 'Z', 'local_c','omega'])
total_time = np.array([])
total_progvar_values = np.array([])
total_mix_frac = np.array([])
total_local_c = np.array([])
total_chi = np.array([])
total_omega = np.array([])
for filename in matching_files_sorted:
    path = os.path.join(results_path,filename)
    data = pd.read_csv(path, sep='\t', skiprows = 1)
    mix_frac = data['Z']
    progvar = data['ProgVar']
    omega = data['ProdRateProgVar [kg/m^3s]']
    progvar_at_z = np.array(data.loc[data['Z'] == z_st, 'ProgVar'])
    second_part = filename.split("_")[1]
    time_str = second_part.split("t")[1]
    time_str = time_str.replace('ms','').strip()
    time_str = time_str.rstrip('.')
    time = float(time_str)
    progvar_values = np.append(progvar_values,progvar_at_z)
    time_values = np.append(time_values,time)
    specific_time = np.ones(len(mix_frac)) * time 
    specific_chi = np.ones(len(mix_frac)) * chi
    specific_progvar = np.ones(len(mix_frac)) * progvar_at_z
    total_time = np.append(total_time, specific_time)
    total_chi = np.append(total_chi, specific_chi)
    total_progvar_values = np.append(total_progvar_values, specific_progvar)
    total_mix_frac = np.append(total_mix_frac, mix_frac)
    total_local_c = np.append(total_local_c, progvar)
    total_omega = np.append(total_omega,omega)


df['time'] = total_time
df['chi'] = total_chi
df['c_st'] = total_progvar_values
df['Z'] = total_mix_frac
df['local_c'] = total_local_c
df['omega'] = total_omega
resultname = (f'./UFPV_data//chi_{chi}_e-01.csv')
df.to_csv(resultname)
```
